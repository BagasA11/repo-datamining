```python
def trim_space(input):
  text = ''
  if type(input) != str:
    input = str(input)
  for i in range(len(input)):
    if input[i] != ' ':
      text += input[i]
  return text

```

```python
# prompt: covert CFR_Harian and RI_Harian to numeric

# Convert 'CFR_Harian' and 'RI_Harian' to numeric, handling errors
def to_numeric(value):
    try:
        return pd.to_numeric(value, errors='coerce')
    except:
        return value

df_covid['CFR_Harian'] = df_covid['CFR_Harian'].apply(to_numeric)
df_covid['RI_Harian'] = df_covid['RI_Harian'].apply(to_numeric)

# Now check the data types again
print(df_covid['CFR_Harian'].dtype)
print(df_covid['RI_Harian'].dtype)

# Further handling of non-numeric values (if any)
df_covid['CFR_Harian'] = pd.to_numeric(df_covid['CFR_Harian'], errors='coerce')
df_covid['RI_Harian'] = pd.to_numeric(df_covid['RI_Harian'], errors='coerce')

# Fill NaN values resulting from the conversion (if any) with 0
df_covid['CFR_Harian'].fillna(0, inplace=True)
df_covid['RI_Harian'].fillna(0, inplace=True)

print(df_covid[['CFR_Harian', 'RI_Harian']])
```
