```python
import pandas as pd
base_path = '/content/drive/MyDrive/KULIAH/smt-5/datamining/assignment-5/'
df = pd.read_csv(base_path + 'merged.csv')
df.head()
```

```python
df = df.set_index('Unnamed: 0')
df.head()
```

```python
# prompt: get mean, median, and standard deviation from df

print(df.mean())
print(df.median())
print(df.std())
```

```python
cv = df.std() / df.mean()
print(cv)
```

```python
daerah = list(df.index)
X = df.iloc[:, :].values
X
```

```python
# prompt: make standarization of X and assign output to new variable

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
# prompt: build k-means model from df with n = 4
# original
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)
```

```python
# standarisasi
kmeans_standarized = KMeans(n_clusters=4, random_state=42)
y_kmeans_standarized = kmeans_standarized.fit_predict(X_scaled)
```

```python
# prompt: create visual diagram of cluster from trained model

import matplotlib.pyplot as plt

# Assuming df has at least two numerical features for plotting
# If not, you'll need to select appropriate features or use dimensionality reduction

plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)

plt.xlabel(df.columns[0])  # Replace with actual feature name
plt.ylabel(df.columns[1])  # Replace with actual feature name
plt.title('KMeans Clustering')
plt.show()
```

```python
# prompt: create cluester visual with standarized kmeans model

# Assuming df has at least two numerical features for plotting
# If not, you'll need to select appropriate features or use dimensionality reduction

plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=y_kmeans_standarized, s=50, cmap='viridis')

centers = kmeans_standarized.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.5)

plt.xlabel(df.columns[0])  # Replace with actual feature name
plt.ylabel(df.columns[1])  # Replace with actual feature name
plt.title('KMeans Clustering (Standardized)')
plt.show()
```
